#### OpenCv Model

```python
face = cv2.imread('DataSets\detect-face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
plt.imshow(face)
```

```python
face_detect_model = cv2.CascadeClassifier('Models/facedetector.xml')
face_detect_model
```

```python
detectface = face_detect_model.detectMultiScale(face, scaleFactor=1.1,minNeighbors=4)
detectface
```

```python
for (x, y, w, h) in detectface:
    cv2.rectangle(face, (x,y), (w+x, h+y), (255,0,0), 10)

plt.imshow(face)
```
